### Creating the various functions needed

```python
#Loading the features of the dataset
news_train = fetch_20newsgroups_vectorized(subset='train',remove=('headers', 'footers', 'quotes'))
X_train, y_train = news_train.data, news_train.target

news_test = fetch_20newsgroups_vectorized(subset='train',remove=('headers', 'footers', 'quotes'))
X_test, y_test = news_test.data, news_train.target
```

```python
#Loading class names
class_names = news_train.target_names
```

```python
def run_multiclass_pca (model, components, x_train, x_test, y_train, y_test):
    
    #Scaling the features
    sc = StandardScaler(with_mean=False)
    X_train = sc.fit_transform(x_train)
    X_test = sc.transform(x_test)
    
    #Using Kernel PCA for dimentionality reduction
    #Setting the first two principal components of the data
    kpca = KernelPCA(n_components=components)
    #Transforming data onto the first two principal components
    x_train_pca = kpca.fit_transform(X_train)
    x_test_pca = kpca.transform(X_test)
    
    #Fitting regression model
    base_model = model
    model_fit = base_model.fit(x_train_pca, y_train)
    predictions = model_fit.predict(x_test_pca)
    
    #Computing for accuracy scores
    training_score = base_model.score(x_train_pca, y_train)
    test_score = base_model.score(x_test_pca, y_test)
    #Creating a confusion matrix
    cm = confusion_matrix(y_test, predictions)
    return model_fit, x_train_pca, training_score, test_score, predictions, cm, components
```

### Checking and visualizing the first two principal components

```python
#Running the first two principal components of the data
kpca2 = KernelPCA(n_components=2)
#Transforming data onto the first two principal components
x_train_pca2 = kpca2.fit_transform(X_train)
```

```python
#Visualising the first vs. second principal component, colored by class
plt.figure(figsize=(15, 15))
mglearn.discrete_scatter(x_train_pca2[:, 0], x_train_pca2[:, 1], y_train)
plt.legend(class_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

### Training classifier models with various dimensions

```python
#Setting the number of components
components = [2,3,4,5,6,8]
#Initializing scores
test_score = []
train_score = []
#Running the models at different dimensions
for i in components:
    classifier = run_multiclass_pca(model=KNeighborsClassifier(n_jobs=-1), components=i, x_train=X_train, x_test=X_test, y_train=y_train, y_test=y_test)
    train_score.append(classifier[2])
    test_score.append(classifier[3])
```

### Visualising the accuracy scores for the models run on different components

```python
plt.figure(figsize=(12, 6))
plt.plot(components, train_score, label="training accuracy",marker='o')
plt.plot(components, test_score, label="test accuracy",marker='p')
plt.title("Accuracy scores of the classification models plotted on number of dimensions")
plt.ylabel('Accuracy')
plt.xlabel('Dimensions')
plt.legend()
```

From the plot of the accuracy scores, it appears that the test accuracy score increases as we increase the dimension. However, the trade of this is the fitting time of the model - I noticed that the higher the dimensions, the longer it took to fit the model.

### Clustering the data into 20 clusters

```python
#Reducing the dimensions to n_components = 2
kpca2 = KernelPCA(n_components=2)
#Transforming data onto the first two principal components
x_train_pca2 = kpca2.fit_transform(X_train)
```

```python
#Clustering the data using the model K-Means clustering (reduced to 2-dimensions)
kmeans = KMeans(n_clusters=20, random_state=10)
kmeans.fit(x_train_pca2)
y_pred = kmeans.predict(x_train_pca2)
plt.scatter(x_train_pca2[:, 0], x_train_pca2[:, 1], c=y_pred, s=60, cmap='Paired')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=60,
marker='^', c=range(kmeans.n_clusters), linewidth=2, cmap='Paired')
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
print("Cluster memberships:\n{}".format(y_pred))
```

### Printing and Checking True Class versus Clustering

```python
#Combining Original Classes and Clustering using the model into one dataframe
Clusters= pd.DataFrame()
Clusters['True Class'] = y_train
Clusters['Cluster Pred'] = y_pred
```

### Checking Confusion matrix as heatmap: Did we recover original clusters?

```python
#Computing for confusion matrix
cm_20 = confusion_matrix(y_train, y_pred)
#Plotting heatmap of confusion matrix per cluster
plt.figure(figsize = (12,6))
sns.heatmap(cm_20, yticklabels=class_names, 
           xticklabels=["cluster {}".format(i) for i in range(20)])
plt.show()
```

Looking at the cluster plot, we were able to nicely cluster the data into 20 cluster using the Kmeans model and reducing the dimension of features to 2.

However, if we look at the the confusion matrix heatmap of the cluster versus original classes, it seems like we did not really recover the original clusters. Ideally, we would have wanted to see light values on each clusters across the labels. This would indicate one cluster belonging to that label.
