```python
#Importing needed libraries
import pandas as pd
import numpy as np
import category_encoders as ce
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelBinarizer
#Reading and loading the data set for Task 1
import zipfile, requests, io
```

```python

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip'
file_name = 'student-mat.csv'
data = pd.read_csv(zipfile.ZipFile(io.BytesIO(requests.get(url, stream=True).content)).open(file_name),sep=';')
```

```python
#Checking the datatypes to determine variable of type objects
data.dtypes
```

```python
#Creating data of object types only
object_data = data.select_dtypes(include=['object']).copy()
object_data.head()
```

```python
#Checking for null data
object_data[object_data.isnull().any(axis=1)]
```

```python
#Using Label Encoder to encode sex, address, famsize, Pstatus
label_enc = LabelEncoder()
columns_le = [ 'sex', 'address', 'famsize', 'Pstatus']
object_data[columns_le] = object_data[columns_le].apply(label_enc.fit_transform)
object_data.head()
```

```python
#Copy a fresh dataset
object_lb = object_data.copy()
#Using LabelBinarizer to encode school
object_lb['school'] = LabelBinarizer().fit_transform(object_lb['school'])
object_lb.head()
```

```python
#Using One Hot Encoder to encode the variables with yes or no answers
pd.get_dummies(object_lb, columns=['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic'], prefix=['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']).head()
```

```python
#Using Label encoding in Pandas by using category values
object_lb['Mjob'] = object_lb['Mjob'].astype('category')
object_lb['Mjob_cat'] = object_lb['Mjob'].cat.codes
object_lb.head()
```

```python
object_lb["Fjob"].value_counts()
```

```python
fjob_nums = {"Fjob" : {"at_home":1,"health":2,"teacher":3,"services":4,"other":5}}
```

```python
object_lb.replace(fjob_nums, inplace=True)
object_lb.head()
```

```python
#Using category encoders to encode reason
#Copying a fresh dataset
object_ce = object_data.copy(deep=True)

# Specify the columns to encode then fit and transform
encoder = ce.sum_coding.SumEncoder(cols=["reason"])
encoder.fit(object_ce, verbose=0)

# Only display the first 8 columns for brevity
encoder.transform(object_ce).head()
```

```python
import pandas as pd
# The file has no headers naming the columns, so we pass header=None
# and provide the column names explicitly in "names"
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
data = pd.read_csv(url, header=None, index_col=False,
names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','income'])
# For illustration purposes, we only select some of the columns
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week',
'occupation', 'income']]
# IPython.display allows nice output formatting within the Jupyter notebook
display(data.head())
```

```python
print(data.income.value_counts())
```

```python
data.dtypes
```

```python
print("Original features:\n", list(data.columns), "\n")
# automatically transforms all columns that have object type (like strings) or are categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
data_dummies = data.copy(deep=True)
data_dummies = pd.get_dummies(data_dummies, columns=["workclass", "education", "gender","occupation",'income'], prefix=["workclass", "education", "gender","occupation",'income'])


print("Features after get_dummies:\n", list(data_dummies.columns))
```

```python
le = LabelEncoder()
le.fit(data['income'])
data['income_rc'] = le.fit_transform(data['income'])
```

```python
data_dummies.head()
```

In this case, we extract only the columns containing featuresâ€”that is, all columns from age to occupation_ Transport-moving. This range contains all the features but not the target.

```python
import numpy as np
features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']
# Extract NumPy arrays
X = np.asarray(features)
y = np.asarray(data["income_rc"])

print("X.shape: {} y.shape: {}".format(X.shape, y.shape))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#Split the train set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

# Train a Logostic regression on X and y
logreg = LogisticRegression(max_iter=10000).fit(X_train, y_train)

print("Test score: {:.2f}".format(logreg.score(X_test, y_test)))
```
